```python
# Make local pygmt of the bardabunga volcano in iceland
import pygmt
fig = pygmt.Figure()
fig.coast(region=[-25, -10, 62, 67], projection='M10c', land='gray', water='skyblue', frame=True)
# Add the volcano with a volcano icon
fig.plot(x=-17.5, y=64.5, style='kvolcano/1.0c', color='red', pen='black', label='Bárðarbunga')
fig.show()
```

```python
# Import pygmt and create a figure with a global map
import pygmt
fig = pygmt.Figure()
# create map of the globe
fig.coast(region='g', projection='W0/15c', land='gray', water='skyblue', frame=True)
# show the map
fig.show()
```

```python
# Get coordinates of seismic stations from the IRIS database
import obspy
from obspy.clients.fdsn import Client
client = Client("IRIS")
# Get the coordinates of all stations in the FDSN network
inventory = client.get_stations(network="II,IU")
```

```python
# Get the coordinates of all stations from the inventory
lats = []
lons = []
for network in inventory:
    for station in network:
        lats.append(station.latitude)
        lons.append(station.longitude)

```

```python
# Create global map with the coordinates of the stations
import pygmt
# Create a figure
fig = pygmt.Figure()
# Create global map
fig.coast(region='g', projection='W0/15c', land='lightgray', water='white', frame=True)
#plot that station location on to the map
fig.plot(x=lons, y=lats, style='t0.2c', fill='red', pen='black')
# show the figure
fig.show()
fig.savefig('global_map.pdf')
```

```python
# Make a matplitlib figure with size 8, 4
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8, 4))
# plot sin and cosine in the same figure
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()
# fig.savefig('sin_cos.pdf')
#show figure
plt.show()
```
